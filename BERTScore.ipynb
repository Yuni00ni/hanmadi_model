```python
from bert_score import score

import pandas as pd
from tqdm import tqdm
import time

for i in tqdm(range(100)):
    time.sleep(0.01)  # 작업이 진행되는 것처럼 보이게 잠깐 멈춤


# 1. 데이터셋 불러오기
df_origin = pd.read_csv('../hanmadi_model/data/cleaned_data.csv')     # 원본 뉴스 데이터
df_summary = pd.read_csv('../hanmadi_model/data/merged_news_data.csv')     # KoBERT가 만든 요약 데이터

# 2. 본문과 요약문 컬럼 추출 (형태는 상황에 맞게 조정)
references = df_origin['본문_clean'].astype(str).tolist()      # 사람이 쓴 뉴스 원문
hypotheses = df_summary['요약문'].astype(str).tolist()   # AI가 만든 요약

# 3. 평가 수행
P, R, F1 = score(hypotheses, references, lang="ko")

# 4. 평균 점수 출력
print(f"[BERTScore 평균]")
print(f"Precision: {P.mean().item():.4f}")
print(f"Recall:    {R.mean().item():.4f}")
print(f"F1 Score:  {F1.mean().item():.4f}")
```
